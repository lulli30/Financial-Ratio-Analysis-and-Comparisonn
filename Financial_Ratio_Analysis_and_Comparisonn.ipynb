```python
df.columns = df.columns.str.strip()
```

```python
key_ratios = [
    'ROA(C) before interest and depreciation before interest',
    'Net Income to Total Assets',
    'Liability to Equity',
    'Interest Coverage Ratio (Interest expense to EBIT)'
]
```

```python
df_non_bankrupt = df[df['Bankrupt?'] == 0]
df_bankrupt = df[df['Bankrupt?'] == 1]
```

```python
# Create the grouped means
group_means = df.groupby('Bankrupt?')[key_ratios].mean()

# Apply a background gradient (Green for high values, Red for low)
# This makes the "Profitability Gap" stand out immediately
group_means.style.background_gradient(cmap='RdYlGn').format("{:.4f}")

print("\n" + "="*80)
print("AVERAGE FINANCIAL RATIOS BY BANKRUPTCY STATUS".center(80))
print("="*80)

# Displaying the dataframe with better spacing
print(group_means.to_string(index=True, justify='center'))

print("="*80)

# Calculate the percentage difference
diff = ((group_means.loc[1] - group_means.loc[0]) / group_means.loc[0]) * 100

# Combine them into a single report table
report = group_means.T # Transpose for easier reading
report.columns = ['Healthy (0)', 'Bankrupt (1)']
report['% Change'] = diff

print("FINANCIAL GAP ANALYSIS")
print(report.to_string())
```

Profitability Gap: Bankrupt companies exhibit significantly lower profitability, with their average Return on Assets (ROA) being nearly $18\%$ lower than non-bankrupt firms, suggesting inefficient asset management or operational struggles long before failure.

Solvency Risks: There is a clear trend of higher financial risk in bankrupt entities, shown by an elevated Liability to Equity ratio and a lower Net Income to Total Assets ratio, which indicates a heavy reliance on debt and an inability to generate surplus value from the asset base.

Predictive Value: While ratios like Interest Coverage show narrow differences in averages, the profitability metrics (ROA and Net Income) serve as much stronger leading indicators for identifying companies at risk of bankruptcy within this dataset.

```python
# 1. Set up the figure and grid
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('Financial Ratio Analysis: Bankrupt vs Non-Bankrupt Companies',
             fontsize=20, fontweight='bold', y=0.98)

# 2. Define colors (Green for healthy, Red/Salmon for Bankrupt)
colors = ['#A1D99B', '#F8766D']

# 3. List of ratios to plot (Adding 'Net Income to Total Assets' to fill the 4th spot)
ratios_to_plot = [
    ('ROA(C) before interest and depreciation before interest', 'ROA Comparison', 'ROA Value'),
    ('Net Income to Total Assets', 'Net Income to Total Assets', 'Ratio Value'),
    ('Liability to Equity', 'Liability to Equity Ratio', 'Liability/Equity'),
    ('Interest Coverage Ratio (Interest expense to EBIT)', 'Interest Coverage Ratio', 'Coverage Ratio')
]

# 4. Loop through axes and ratios to build the plots
for i, (column, title, ylabel) in enumerate(ratios_to_plot):
    # Determine the current axis (0,0), (0,1), (1,0), (1,1)
    ax = axes[i // 2, i % 2]

    # Create the boxplot
    # showfliers=False is recommended because financial outliers can hide the actual box
    bp = ax.boxplot([df_non_bankrupt[column], df_bankrupt[column]],
                    tick_labels=['Non-Bankrupt', 'Bankrupt'],
                    patch_artist=True,
                    widths=0.6,
                    showfliers=False)

    # Apply styling to the boxes
    for patch, color in zip(bp['boxes'], colors):
        patch.set_facecolor(color)
        patch.set_edgecolor('#444444')
        patch.set_alpha(0.8)

    # Style the median lines (the line inside the box)
    for median in bp['medians']:
        median.set(color='black', linewidth=2)

    # Add labels and formatting
    ax.set_title(title, fontsize=14, fontweight='bold', pad=10)
    ax.set_ylabel(ylabel, fontsize=12)
    ax.grid(True, linestyle='--', alpha=0.5, axis='y')
    ax.set_facecolor('#fdfdfd') # Very light grey background for a clean look

# 5. Finalize and show
plt.tight_layout(rect=[0, 0, 1, 0.96]) # Adjust for the main title
plt.show()
```

1. The Interquartile Range (IQR) Gap: "Performance Segregation"
The Visual: Look at the ROA and Net Income to Total Assets plots. You will see that the entire box for "Bankrupt" companies is physically lower than the box for "Healthy" companies.

when the boxes (the middle 50% of data) do not overlap, it indicates a highly significant difference. Most bankrupt companies are performing worse than even the bottom 25% of healthy companies. There is almost no "middle ground" if a company’s ROA falls into that lower box, it is in a "danger zone" where very few healthy companies exist.

2. Box Height: "Financial Instability"
The Visual: Compare the vertical height (the "tallness") of the boxes. Often, the Bankrupt box is taller than the Healthy box.

A taller box means a higher standard deviation (more spread). Healthy companies are very consistent; they all hover around a similar "safe" profit margin. Bankrupt companies are erratic—some are losing a little, some are losing a lot, and some are drowning in debt. This "vertical stretch" in the boxplot is a visual signature of financial volatility.

3. The Median Shift: "Operational Failure"
The Visual: Look at the horizontal line inside the boxes (the Median).

This is the "typical" company's performance. In the Liability to Equity plot, the bankrupt median is shifted upward. This confirms that the typical failing company is carrying significantly more debt than the typical healthy company. It’s not just a few bad apples; the entire group has shifted toward a high-risk debt structure.

4. Whisker Reach: "The Best of the Worst"
The Visual: Look at the top "whisker" (the line extending upward) of the Bankrupt group.

This represents the "best-performing" companies within the failing group.In the ROA plot, notice that even the "top" whisker of the bankrupt group often doesn't reach the Median of the healthy group. This means even the "best" bankrupt company is still performing worse than an average healthy company.

5. Summary of the Visual Evidence
Separation = Predictability: Because the boxes for ROA are so far apart, ROA is a Strong Predictor.

Overlap = Low Predictability: Because the boxes for Interest Coverage are closer together and overlap more, it is a Weak Predictor. The boxplot shows you visually why you should trust the ROA data more than the Interest Coverage data when trying to spot a failing business.

```python
print("\n" + "="*80)
print("PERCENTAGE DIFFERENCES (Bankrupt vs Non-Bankrupt):")
print("="*80)

for ratio in key_ratios:
    non_bankrupt_mean = df_non_bankrupt[ratio].mean()
    bankrupt_mean = df_bankrupt[ratio].mean()

    if non_bankrupt_mean != 0:
        pct_diff = ((bankrupt_mean - non_bankrupt_mean) / non_bankrupt_mean) * 100
    else:
        pct_diff = float('inf')

    print(f"\n{ratio[:50]}...:")
    print(f"  Non-Bankrupt: {non_bankrupt_mean:.4f}")
    print(f"  Bankrupt:     {bankrupt_mean:.4f}")
    print(f"  % Difference: {pct_diff:+.2f}%")
```

```python
ratios_for_bar = key_ratios[:3]
x_labels = ['ROA', 'Net Income/Assets', 'Liability/Equity']

non_bankrupt_means = [df_non_bankrupt[ratio].mean() for ratio in ratios_for_bar]
bankrupt_means = [df_bankrupt[ratio].mean() for ratio in ratios_for_bar]

x = np.arange(len(x_labels))
width = 0.35

plt.bar(x - width/2, non_bankrupt_means, width, label='Non-Bankrupt',
        color='skyblue', edgecolor='black')
plt.bar(x + width/2, bankrupt_means, width, label='Bankrupt',
        color='salmon', edgecolor='black')

plt.xlabel('Financial Ratios')
plt.ylabel('Average Value')
plt.title('Average Financial Ratios: Bankrupt vs Non-Bankrupt Companies')
plt.xticks(x, x_labels)
plt.legend()
plt.grid(True, alpha=0.3, axis='y')

#Add value labels on bars
for i in range(len(x_labels)):
    plt.text(i - width/2, non_bankrupt_means[i] + 0.001,
             f'{non_bankrupt_means[i]:.3f}', ha='center', va='bottom', fontsize=9)
    plt.text(i + width/2, bankrupt_means[i] + 0.001,
             f'{bankrupt_means[i]:.3f}', ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()

#Summary statistics
print("\n" + "="*80)
print("SUMMARY STATISTICS:")
print("="*80)

print("\nNon-Bankrupt Companies:")
print(f"Count: {len(df_non_bankrupt)}")
for ratio in key_ratios:
    mean_val = df_non_bankrupt[ratio].mean()
    std_val = df_non_bankrupt[ratio].std()
    print(f"{ratio[:30]}...: Mean={mean_val:.4f}, Std={std_val:.4f}")

print("\nBankrupt Companies:")
print(f"Count: {len(df_bankrupt)}")
for ratio in key_ratios:
    mean_val = df_bankrupt[ratio].mean()
    std_val = df_bankrupt[ratio].std()
    print(f"{ratio[:30]}...: Mean={mean_val:.4f}, Std={std_val:.4f}")
```
